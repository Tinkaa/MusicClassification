```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
```

```python
xtrain = np.loadtxt('Data/train_data.csv', delimiter=',')
ytrain = np.loadtxt('Data/train_labels.csv', delimiter=',')
xtest = np.loadtxt('Data/test_data.csv', delimiter=',')
```

```python
# convert integers to dummy variables (i.e. one hot encoded)
dummy_y = np_utils.to_categorical(ytrain)
#since our classes start at 1 and not at 0
dummy_y=np.delete(dummy_y,0,1)
```

```python
dummy_y
```

```python
# define baseline model
def baseline_model():
    # create model
    model = Sequential()
    model.add(Dense(8, input_dim=xtrain.shape[1], activation='relu'))
    model.add(Dense(10, activation='softmax'))
    # Compile model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
#build the model
estimator = KerasClassifier(build_fn=baseline_model, epochs=10, batch_size=50, verbose=0)
```

```python
#use kfold cross-validation
kfold = KFold(n_splits=10, shuffle=True)
```

```python
#compute the results
results = cross_val_score(estimator, xtrain, dummy_y, cv=kfold)
print("Baseline: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

```python
#fit the model and predict
estimator.fit(xtrain,ytrain)
ytrainpred=estimator.predict(xtrain)
```
