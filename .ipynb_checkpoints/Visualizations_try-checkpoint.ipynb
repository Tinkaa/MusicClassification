```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
%matplotlib inline

from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.manifold.t_sne import TSNE
from sklearn.neighbors.classification import KNeighborsClassifier
```

```python
x = np.loadtxt('Data/train_data.csv', delimiter=',')
y = np.loadtxt('Data/train_labels.csv', delimiter=',')
x_predict = np.loadtxt('Data/test_data.csv', delimiter=',')
scaler = StandardScaler()
x = scaler.fit_transform(x)
x_predict = scaler.fit_transform(x_predict)
X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.9)
```

```python
#use PCA to decrease features to two for visualization
pca = PCA(n_components=2)
pca.fit(X_train)
```

```python
#same but for Linear Discriminant Analysis instead of PCA
lda = LDA(n_components=2) #2-dimensional LDA
lda_transformed = pd.DataFrame(lda.fit_transform(X_train, y_train))
```

```python
#again but now with T-sne. This time we try to plot the decision boundary from the SVM
X_Train_embedded = TSNE(n_components=2).fit_transform(X_train)
C_ = 2.9
clf = svm.SVC(decision_function_shape='ovo', random_state=42, C=C_)
clf.fit(X_train, y_train)
y_predicted = clf.predict(X_train)
```

```python
#try to get some info on decision boundary, e.g. is it way too blobby? 
#not much success, is just chaos.. 
#copied from https://stackoverflow.com/questions/37718347/plotting-decision-boundary-for-high-dimension-data
# create meshgrid
resolution = 100 # 100x100 background pixels
X2d_xmin, X2d_xmax = np.min(X_Train_embedded[:,0]), np.max(X_Train_embedded[:,0])
X2d_ymin, X2d_ymax = np.min(X_Train_embedded[:,1]), np.max(X_Train_embedded[:,1])
xx, yy = np.meshgrid(np.linspace(X2d_xmin, X2d_xmax, resolution), np.linspace(X2d_ymin, X2d_ymax, resolution))

# approximate Voronoi tesselation on resolution x resolution grid using 1-NN
background_model = KNeighborsClassifier(n_neighbors=1).fit(X_Train_embedded, y_predicted) 
voronoiBackground = background_model.predict(np.c_[xx.ravel(), yy.ravel()])
voronoiBackground = voronoiBackground.reshape((resolution, resolution))

#plot
plt.contourf(xx, yy, voronoiBackground)
X_scat=X_Train_embedded[:100]
y_scat=y_predicted[:100]
plt.scatter(X_scat[:,0], X_scat[:,1], c=y_scat)
plt.colorbar()
plt.show()
```
